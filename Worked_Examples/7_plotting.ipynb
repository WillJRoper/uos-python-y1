## 3.

```python
def gauss(x, mu=0, sigma=1):
    """
    Calculate the value of a normal distriution at x.

    Arguments are set to a default but can be anything.

    Args:
        x (float)
            Value to evaluate at.
        mu (float)
            The mean.
        sigma (float)
            The standard deviation
    """

    # Define the normalisation
    norm = 1 / (sigma * np.sqrt(2 * np.pi))

    # Define the exponent
    exponent = ((x - mu) / (2 * sigma)) ** 2

    return norm * np.exp(-exponent)

# Define x values
x = np.arange(-3.2, 3.2, 0.01)

# Calculate y values (we can use the default values for mu and sigma)
y = gauss(x)

# Plot the results
plt.plot(x, y)

# Label the axes
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14)

# Add a title
plt.title("Gaussian", fontsize=14)

plt.show()
```

```python
import random


# Create the random x values between -5 and 5
i = 0
x1 = []
x2 = []
while i < 20:
    x1.append(random.random() * 10 - 5)
    x2.append(random.random() * 10 - 5)
    i += 1

# Convert lists to arrays
x1 = np.array(x1)
x2 = np.array(x2)

# Create the high resolution x values
x = np.arange(-5, 5, 0.05)

# Set up the first figure
plt.figure(1)

# Plot the scatter
plt.scatter(x1, np.sin(2*x1 + 1))

# Plot the line
plt.plot(x, np.sin(2*x + 1))

# Label the axes
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14)

# Add a title
plt.title("$\sin(2x+1)$", fontsize=14)

# Limit the axes
plt.xlim(-5, 5)
plt.ylim(-1.2, 1.2)

# Set up the second figure
plt.figure(1)

# Plot the scatter
plt.scatter(x2, np.cos(2*x2 + 1))

# Plot the line
plt.plot(x, np.cos(2*x + 1))

# Label the axes
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14)

# Add a title
plt.title("$\cos(2x+1)$", fontsize=14)

# Limit the axes
plt.xlim(-5, 5)
plt.ylim(-1.2, 1.2)

plt.show()
```

## 1.

```python
def coin_flip():
    """
    Flip a coin.

    Returns:
        bool
            True for heads, False for tails.
    """

    # Get a random number
    num = random.random()

    # If the number is greater or equal to 0.5 its heads
    if num >= 0.5:
        return True
    
    # Otherwise, tails
    return False

# Define counters for number of heads and tails
nheads = 0
ntails = 0

# Flip the coin 100 times and count the number of heads and tails
while nheads + ntails < 100:
    heads = coin_flip()

    # Increment the correct counter
    if heads:
        nheads += 1
    else:
        ntails += 1

print(f"There were {nheads} heads and {ntails} tails")
```
